```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
# Import csv as a DataFrame
csv_path = Path("2.15excel.csv")
poverty_df = pd.read_csv(csv_path)
poverty_df.head()
```

```python
# Drop top three N/A rows
poverty_df = poverty_df.drop(poverty_df.index[[0,1,2]])
poverty_df.head()
```

```python
# Set the first row as the header
new_header = poverty_df.iloc[0]
poverty_df = poverty_df[1:]
poverty_df.columns = new_header
poverty_df.head()
```

```python
# Create a new dataframe for future use of retrieving the most recent value
cleaning_df = poverty_df.iloc[:, 4:]
cleaning_df.head()
```

```python
# Retrieve the most recent value and append dataframe
def f(x):
    if x.last_valid_index() is None:
        return 0
    else:
        return x[x.last_valid_index()]

cleaning_df['<$2.15'] = cleaning_df.apply(f, axis=1)
cleaning_df.head()
```

```python
# Concatenate dataframes
poverty_df_clean = pd.concat([poverty_df['Country Name'], poverty_df['Country Code'],
                              cleaning_df['<$2.15']], axis=1)
poverty_df_clean.head(10)
```

```python
# Save the data to a CSV file without the index
poverty_df_clean.to_csv("Resources/poverty_df.csv", index=False)
```

```python
# We want to only include latitude and longitude for the countries that we have
csv_path = Path("Resources/corruption_data.csv")
corruption_df = pd.read_csv(csv_path)
corruption_df.head()
```

```python
# Convert the 'Code' column to list to use in verification loop
countries = corruption_df['Code'].tolist()
len(countries)
```

```python
# If the country in the poverty dataset does not have a corresponding latitude/longitude, drop it
for index, row in poverty_df_clean.iterrows():
    if row["Country Code"] not in countries:
        poverty_df_clean.drop(index, inplace=True)
        
len(poverty_df_clean)
```

```python
# Check for duplicates
poverty_df_clean.drop_duplicates(subset=['Country Code'])
len(poverty_df_clean)
```

```python
# Save the data to a CSV file without the index
poverty_df_clean.to_csv("Resources/poverty_df_clean.csv", index=False)
```
