```python
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_csv('gender-gap-education-levels.csv')
data_df.head()
```

```python
# Grab 2020 year data
data_df.rename(columns={'Entity': 'Country'}, inplace=True)
data2020_df = data_df[data_df.Year == 2020]
data2020_df
```

```python
# Selecting only tertiary educaiton columns for 2020
selected_columns = data2020_df[['Code','Country','Combined gross enrolment ratio for tertiary education, female', 'Combined gross enrolment ratio for tertiary education, male']]
selected_columns
```

```python
selected_columns.info()
```

```python
year_2020_df = selected_columns.set_index('Country')
year_2020_df 
```

```python
countries_df = pd.read_csv('corruption_data.csv', index_col= 'Code' )
countries_df 
```

```python
columns_drop = countries_df[['Rank','Corruption Index Score','latitude','longitude']]
c_df = countries_df.drop(columns=columns_drop)
c_df
```

```python
merge_df = pd.merge(c_df, year_2020_df, on='Code', how='outer')
merge_df
```
